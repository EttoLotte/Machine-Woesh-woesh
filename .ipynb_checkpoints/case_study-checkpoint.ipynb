# Exercise 1.6

```python
def calculate_features(image_path):
    """
    Calculates features based on the given image.
    
    Parameters:
        image_path (str): Path to the input image.
        
    Returns:
        dict: A dictionary containing all calculated features.
    """
    image = io.imread(image_path, as_gray=True)  
    edges = feature.canny(image)  
    non_zero_indices = np.argwhere(edges) 
    height = np.max(non_zero_indices[:, 0]) - np.min(non_zero_indices[:, 0]) + 1
    width = np.max(non_zero_indices[:, 1]) - np.min(non_zero_indices[:, 1]) + 1
    max_pixels_per_column = np.max(np.sum(edges, axis=0))
    max_pixels_per_row = np.max(np.sum(edges, axis=1))
    total_pixels = np.sum(edges)
    labeled_image, num_edges = measure.label(edges, connectivity=2, return_num=True)
    total_image_pixels = edges.size  
    edge_density = total_pixels / total_image_pixels
    aspect_ratio = height / width
    features = {
        "Number of edges": num_edges,
        "Height of digit": height,
        "Width of digit": width,
        "Maximum number of pixels per column": max_pixels_per_column,
        "Maximum number of pixels per row": max_pixels_per_row,
        "Total number of pixels": total_pixels,
        "Edge density": edge_density,
        "Aspect ratio": aspect_ratio,
    }
    return features

image_path = 'Reduced Trainging data/8/4857.jpg'
features = calculate_features(image_path)

for feature_name, value in features.items():
    if isinstance(value, float):
        print(f"{feature_name}: {value:.2f}")
    else:
        print(f"{feature_name}: {value}")
```

# Exercise 1.8

# Functies

```python
def calculate_features(image_path):
    """
    Calculates features for a given image (Canny edge detection).
    """
    image = io.imread(image_path, as_gray=True)  
    if image.ndim != 2:
        raise ValueError("De afbeelding moet 2 dimensies hebben (grijswaarden).")
    edges = feature.canny(image)
    non_zero_indices = np.argwhere(edges)  
    height = non_zero_indices[:, 0].max() - non_zero_indices[:, 0].min()
    width = non_zero_indices[:, 1].max() - non_zero_indices[:, 1].min()
    max_pixels_per_column = np.max(np.sum(edges, axis=0))
    max_pixels_per_row = np.max(np.sum(edges, axis=1))
    total_pixels = np.sum(edges)
    labeled_image, num_edges = measure.label(edges, connectivity=2, return_num=True)
    total_image_pixels = edges.size  
    edge_density = total_pixels / total_image_pixels
    aspect_ratio = height / width
    return {
        "Number of edges": num_edges,
        "Height of digit": height,
        "Width of digit": width,
        "Maximum number of pixels per column": max_pixels_per_column,
        "Maximum number of pixels per row": max_pixels_per_row,
        "Total number of pixels": total_pixels,
        "Edge density": edge_density,
        "Aspect ratio": aspect_ratio,
    }

def calculate_row_vector(image_path):
    """
    Bereken de rijvector van de afbeelding door de pixelwaarden in een 1D array te stoppen.
    """
    image = io.imread(image_path, as_gray=True)
    if image.ndim != 2:
        raise ValueError("De afbeelding moet 2 dimensies hebben.")
    row_vector = image.reshape(-1)
    
    return row_vector
```

```python
# nog van Exercise 1.8
Trainging_path = 'Reduced Trainging data'
Traingingfiles = []
for i in range(10):
    Trainging_path_i = os.path.join(Trainging_path, str(i))  
    if os.path.exists(Trainging_path_i):
        Trainging_files_i = [os.path.join(Trainging_path_i, f) for f in os.listdir(Trainging_path_i) if f.endswith('.jpg')]
        Traingingfiles.extend(Trainging_files_i)
    else:
        print(f"Pad niet gevonden: {Trainging_path_i}")
```

```python
def process_dataset_from_files(files, base_path=None):
    feature_list = []  
    vector_list = []   
    labels = []       
    for file_path in files:  # Hier gebruik je het volledige pad
        if file_path.endswith(".jpg"):
            # Het label is de naam van de submap ('0' tot '9')
            label = os.path.basename(os.path.dirname(file_path))  
            try:
                features = calculate_features(file_path)
                row_vector = calculate_row_vector(file_path)
                feature_list.append(features)
                vector_list.append(row_vector)
                labels.append(label)
            except Exception as e:
                print(f"Fout bij het verwerken van {file_path}: {e}")

    return feature_list, vector_list, labels

train_features, train_vectors, train_labels = process_dataset_from_files(Traingingfiles)
train_vectors = np.array(train_vectors)
train_labels = np.array(train_labels)

# Split in training (80%) en validatie (20%) sets
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    train_vectors, train_labels, test_size=0.2, random_state=42
)

print(f"Training set size: {X_train.shape}, Validation set size: {X_val.shape}")
```
